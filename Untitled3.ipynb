On commence par importer la libraire TensorFlow qui fourni des fonctions pour la création de système de neurones et la librairie Numpy qui étend les fonctions mathématiques

```python
from keras import *
from tensorflow.keras import layers, models
import numpy as mp
```

on crée une instance du model de l'IA

```python
model = models.Sequential()
model.add(layers.Dense(units=3, input_shape=[1]))
model.add(layers.Dense(units=64))
model.add(layers.Dense(units=1))
```

On commence à spécifier les entrées X et les sorties Y pour les 10 premiers nombres

```python
X = np.array([0, 1, 2, 3, 4, 5])
Y = np.array([0, 1, 4, 9, 16, 25])
```

On compile le model avec des optimisateurs

```python
model.compile(loss='mean_squared_error', optimizer='sgd')
```

On entraine le model avec 500 estimations et vérifications

```python
model.fit(x=X, y=Y, epochs=1000)
```

On essaye de prédire le carré de 10

```python
model.predict([10])
```
